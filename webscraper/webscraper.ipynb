```python
import requests
from bs4 import BeautifulSoup

url = "http://quotes.toscrape.com/"

response = requests.get(url)
if response.status_code == 200:
    html_content = response.text
    soup = BeautifulSoup(html_content, "html.parser")
    
    # Get the first quote without using a loop
    first_quote = soup.find_all("span", class_="text")[0].get_text()
    first_author = soup.find_all("small", class_="author")[0].get_text()
    
    print("First Quote:")
    print("="*50)
    print(f"{first_quote} — {first_author}")
else:
    print("Failed to fetch webpage")
```
