```python
print("üòäWelcome To Python Quiz!")
print("Answer The Following Qs By Typing a, b, c")
print("GOODLUCK!üòÉ")

questions = [
 {
    "question" : "Which is the correct keyword to define a python function?",
    "options" : {"a": "def", "b":"keyword", "c": "function"},
    "correct_answer" : "a"
},
 {

    "question" : "Which symbol is used for making comments in python?",
    "options" : {"a": "!", "b":"#", "c": "//"},
    "correct_answer" : "b"

},
 {

    "question" : "Which data type in python is used to store True/Flase?",
    "options" : {"a": "string", "b":"Boolean", "c": "Variable"},
    "correct_answer" : "b"
},
 {

    "question" : "What is meant by type casting in python?",
    "options" : {"a": "changing type of datatypes", "b":"creating datatypes", "c": "none of above"},
    "correct_answer" : "a"
},
 {
    
    "question" : "Which loops are mostly used in python?",
    "options" : {"a": "for loop", "b":"while loop", "c": "both a & b"},
    "correct_answer" : "c"
}
]

score = 0

for Q in questions:
    print(Q["question"])

    for key, value in Q["options"].items():
        print(f" {key}) {value}")

    user_answer = input("Enter Your Answer!").lower()

    if user_answer == Q["correct_answer"]:
        print("‚úÖcorrect answer!")
        score +=1

    else: 
        print("‚ùåwrong answer!")


print(f"your final score is: {score}/{len(questions)}")
```
