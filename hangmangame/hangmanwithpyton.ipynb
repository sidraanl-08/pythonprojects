## Step1: Setup basis of game 

### 1. we need a list of words that computer will guess
### 2. then we need a random choice for computer to choose between list of words 
### 3. then we need an input function that asks users for letters to guess the word 
### 4. then we will store the guesses of the user in a list
### 5. we can also hardcode our attempts upto 8 attempts to make the game more interesting

## Step2: Make loop of the game

### 1. keep asking user for letters untill they run out of attempts
### 2. ask user for a letter
### 3. we can also check if the letter was already guessed
### 4. we can add user guesses in the empty list as well 
### 5. we can also check if the letter is in the computer choice 
### 6. decrease total attempts in the end 
### 7. also print game over at the end of loop

```python
import random 

words = ["apple", "sidraa", "kiwi", "python", "datascientist", "hangman"]

computer_choice = random.choice(words)

total_attempts = 8

user_guesses = []

while total_attempts > 0:
    print("Attempts left: ", total_attempts)
    user_choice = input("give a letter to guess the word!").lower()

    if user_choice in user_guesses:
        print("choose another letter ")
        continue 

    user_guesses.append(user_choice)

    if user_choice in computer_choice:
        print("‚úÖ correct guess!")

    else:
        print("‚ùå wrong guess!")
        total_attempts -= 1

if total_attempts == 0:
    print("üíÄ GAME OVER! The Word Was: " , computer_choice)
```
